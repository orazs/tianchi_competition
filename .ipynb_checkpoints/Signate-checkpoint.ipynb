# Problem definition
Predicting the health of trees is extremely important in urban greening and forest protection efforts.
Participants will take on the challenge of **predicting the health status of trees** by building analytical models based on quantitative and qualitative data such as tree size, potential problems, and surrounding environment. See more [details](https://signate.jp/competitions/1247)
Unhealthy trees pose a variety of problems to cities and communities. These include a decline in urban aesthetics, reduced capacity to adapt to climate change, and loss of biodiversity. Additionally, it should be considered that unhealthy trees have a high risk of falling, which could cause damage to people's lives and property. Environmental agencies and city managers use methods such as periodic monitoring, public education, implementation of advanced monitoring systems, and early response measures to predict and manage tree health.

```python
import pandas as pd
import numpy as np

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder,Normalizer,StandardScaler,MinMaxScaler,TargetEncoder
from category_encoders.count import CountEncoder
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split, cross_val_score, RandomizedSearchCV,cross_validate
from imblearn.over_sampling import RandomOverSampler

from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import f1_score, classification_report, confusion_matrix
```

# Ingest

```python
nta_extra=pd.read_csv("data/nta_extra.csv")
def prepare_data(train):
    train = train.rename(columns={"Unnamed: 0":"X"})
    train=(train
     .assign(
             problems_parced=train.problems
             .fillna("Notdefined")
             .str.replace(r'(Lights|Grates|Other|Rope)',lambda x: "{0}".format(x.group(1).lower()),regex=True  ) 
             .str.replace(r'([A-Z]+)', r' problems_\1', regex=True)
             .str.strip()
             .str.split(' '),
             problems_cnt = lambda x: x['problems_parced'].str.len(),
             created_month = lambda x: pd.to_datetime(x['created_at']).dt.strftime("%B")
            )
            )
    
    train.loc[pd.isna(train.problems),'problems_cnt']=0
    
    train = pd.concat(
        [train,
        (train
         .assign(
                value =1,
                )
         .explode('problems_parced')
         .pivot(columns = 'problems_parced', values='value')
         .fillna(0)
        )
        ],axis=1)

    train=train.merge(nta_extra, how='left')
    
    return train[["X", 'tree_dbh', 'curb_loc',  'steward', 'guards',
       'sidewalk', 'user_type', 'spc_latin', 'nta',
       'nta_name',  'boroname','zip_city','health', 'problems_cnt',
       'created_month', 'problems_Branchlights', 'problems_Branchother',
       'problems_Metalgrates', 'problems_Notdefined', 'problems_Rootother',
       'problems_Sneakers', 'problems_Stones', 'problems_Trunklights',
       'problems_Trunkother', 'problems_Wiresrope','population_per_tree','area_per_tree']]


    
```

# Transform

```python
train = pd.read_csv("data/train.csv")
train = prepare_data(train)

cat_pipe = Pipeline([
    ('oh',OneHotEncoder(max_categories=10,handle_unknown='ignore',sparse_output=False))
])
    
num_pipe = Pipeline([
    ('impute', SimpleImputer(missing_values=np.nan,strategy='median'))
])

transform = ColumnTransformer(remainder='drop',
                              transformers=
    [
      ('category_pipe',cat_pipe,['curb_loc', 'steward', 'guards','sidewalk', 'user_type', 'spc_latin', 
                                  'boroname','created_month']),
      ("num_pipe",num_pipe,['tree_dbh','problems_cnt','population_per_tree','area_per_tree'])
    ]
).set_output(transform='pandas')


pipe = Pipeline([
    ('tr',transform),
    ("rf",RandomForestClassifier(class_weight='balanced'))
])

params={
    "rf__n_estimators":[150,200,250],
    "rf__max_depth":[5,10,15]
}



x_train,x_test, y_train, y_test = train_test_split(train.drop(["X","health"],axis=1), train['health'], test_size=0.3)

rsearch = RandomizedSearchCV(pipe,param_distributions=params,cv=5,n_jobs=-1,return_train_score=True,scoring="f1_macro")
rsearch.fit(x_train, y_train)
print("best f1 score:{0}".format(rsearch.best_score_))

scores = cross_validate(rsearch.best_estimator_,x_train,y_train, scoring="f1_macro",cv=5,n_jobs=-1,return_train_score=True)
print(scores)
```

# Validation

```python
from plotnine import labs
fi=pd.DataFrame({'name':transform.fit_transform(x_train).columns,
              'val':rsearch.best_estimator_[1].feature_importances_}
            ).sort_values(by='val',ascending=False)
fi['name']=fi['name'].apply(lambda x: x.split('__')[1:][0])

(
ggplot(fi[:20], aes('reorder(name,val)','val'))+
geom_bar(stat='identity')+
coord_flip()+
labs(title='Top 20 most important features')+
theme_minimal()
)
```

```python
y_pred = rsearch.best_estimator_.predict(x_test)
print(confusion_matrix(y_pred, y_test))
print(f1_score(y_pred, y_test,average='macro'))
print(classification_report(y_pred,y_test))
```

# Submission

```python
submission = pd.read_csv("data/test.csv")
submission['health']=0
submission = prepare_data(submission)
submission['health'] = rsearch.best_estimator_.predict(submission.drop(["X","health"],axis=1))
submission[['X','health']].to_csv("data/submission.csv",header=False,index=False)
```

# Extra
Original train dataset has been extended by several new features. Extra data is publicly available:
* [NTA Demographics](https://www.nyc.gov/assets/dfta/downloads/pdf/reports/Demographics_by_NTA.pdf) (population)
* [NTA Tree stats](https://data.cityofnewyork.us/Environment/2015-Street-Tree-Census-Tree-Data/uvpi-gqnh) (alive_trees, total_trees etc)
* [NTA Area](https://data.cityofnewyork.us/City-Government/NTA-map/d3qk-pfyz) (sq nta area)

```python
nta_extra=(pd.read_csv("data/extra/nta_population.csv")
  .loc[lambda x:x['Year']==2010,['NTA Name', 'Population',"NTA Code"]]
  .rename(columns={"NTA Name":"nta_name","Population":"population","NTA Code":"nta"})
      .merge(
          pd.read_csv("data/extra/nta_area.csv")
              .assign(shape_area_sqkm=lambda x:x['shape_area']/10),
          how="left"
      )
     .merge(
         pd.read_csv("data/extra/nta_tree_stats.csv"),
         left_on="nta_name",
         right_on="nta", how='left'
     )[['nta_name','shape_area_sqkm','population','%_utc_parkland','cnt_trees_2016','cnt_of_alive_trees','%_trees_greater_6_dbh', 'cnt_high_scoring_sidewalks',
           '%_trees_greater_30_dbh', 'cnt_harmful_guards']]
 .replace("(%|,)","",regex=True)
)
nta_extra[nta_extra.columns.drop("nta_name")] = nta_extra[nta_extra.columns.drop("nta_name")].apply(pd.to_numeric, errors='coerce')
nta_extra=(
    nta_extra
     .assign(population_per_tree=lambda x:x['population']/x['cnt_trees_2016'],
            area_per_tree = lambda x: x['shape_area_sqkm']/x['cnt_trees_2016'])
)
nta_extra.to_csv("data/nta_extra.csv",index=False)
```
